```python
import nfl_data_py as nfl
import pandas as pd
```

```python
def get_seasonal_data(year):
    year_list = []
    year_list.append(int(year))
    df = nfl.import_seasonal_data(year_list)
    id_df = nfl.import_ids()
    id_df = id_df[['gsis_id','name' , 'position', 'age']]
    df = pd.merge(id_df, 
                    df,  
                     left_on = 'gsis_id',
                     right_on = 'player_id',
                     how ='right')
    # cols = df.columns.tolist()
    # cols = cols[-1:] + cols[:-2]
    # df = df[cols]
    df = df.sort_values('name')

    return df.drop(columns = ['gsis_id', 'season_type'])

latest_season = get_seasonal_data(2023)
latest_season.sample(5)
```
